<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig,ax = plt.subplots(1,3, sharex = True, sharey = True,figsize=(20,8))
ax[0].plot(x, tan, color = 'b', label = 'tanX')
ax[1].plot(x, sin, color = 'r', label = 'sinX')
ax[2].plot(x, cos, color = 'g', label = 'cosX')
fig.legend(loc = 'upper center')
ax[0].set_title('Grafik tanX')
ax[1].set_title('Grafik sinX')
ax[2].set_title('Grafik cosX')
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(13,8))
ax.plot(x,y)
plt.title('Grafik cosX')

ins1 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins1.plot(x, y4, color = 'black')
ins1.plot(x, y, color = 'blue')
ins1.set_title(label = 'grafik cosX ^4')

ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2.plot(x, y5, color = 'y')
ins2.plot(x,y, color = 'b')
ins2.set_title(label = 'grafik cosX ^5')

ins3 = ax.inset_axes([-0.1, -0.2, 0.3, 0.3])
ins3.plot(x,y2, color = 'red')
ins3.plot(x,y, color = 'blue')
ins3.set_title(label = 'grafik cosX ^2')

ins4 = ax.inset_axes([0.9, -0.2, 0.3, 0.3])
ins4.plot(x,y3,color = 'green')
ins4.plot(x,y,color = 'blue')
ins4.set_title(label = 'grafik cosX ^3')

plt.show()
```
